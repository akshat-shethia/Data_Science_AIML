### Introduction to MatplotLib

```python
%matplotlib inline 
#Tells the nb to open the graphs within the nb itself and not in external websites
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = np.array([1,2,3,4])
y = np.array([5,6,7,8])
plt.plot(x,y);
```

### Different Methods of creating plots 

```python
# Method 1
fig = plt.figure() #Creates the skeleton
ax = fig.add_subplot() #adds some axes
plt.show();
```

```python
# Method 2
fig = plt.figure() #Creates the skeleton
ax = fig.add_axes([1,1,1,1]) #adds some axes
ax.plot(x,y)
plt.show();
```

```python
# Method 3 (RECOMMENDED)
fig,ax = plt.subplots()
ax.plot(x,[50,75,150,250]); # Add some Data
type(fig), type(ax)
```

### Matplotlib Example Workflow

```python
# 0) Import Matplotlib and get it ready to work in a jupyter notebook
%matplotlib inline
import matplotlib.pyplot as plt

# 1) Prepare Data
x = [1,2,3,4]
y = [11,20,37,41]

# 2) Setting up the Plot
fig,ax = plt.subplots(figsize = (5,5)) # ( Width, Height )

# 3) Plot axes
ax.plot(x,y)

# 4) Customising the plot
ax.set(title = "Simple Plot",
       xlabel = "X-Axis",
       ylabel = "Y-Axis" )

# 5) Saving the plot as an img
fig.savefig("C:/miniconda/MINICONDA_WORKSPACE/MatPlotLib_Section7/sample-plot.png")
```

### Making Figures With Numpy Arrays

```python
array = np.linspace(0,10,100) #Generates 100 numbers between 0 and 10 ( evenly spaced )
fig,ax = plt.subplots()
ax.plot(array,array**2);
```

```python
# To Get a Scatter Plot
fig,ax = plt.subplots()
ax.scatter(array, np.exp(array));
```

```python
fig,ax = plt.subplots()
ax.scatter(array,np.sin(array))
```
