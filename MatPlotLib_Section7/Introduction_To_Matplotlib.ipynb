### Different Methods of creating plots 

```python
# Method 1
fig = plt.figure() #Creates the skeleton
ax = fig.add_subplot() #adds some axes
plt.show();
```

```python
# Method 2
fig = plt.figure() #Creates the skeleton
ax = fig.add_axes([1,1,1,1]) #adds some axes
ax.plot(x,y)
plt.show();
```

```python
# Method 3 (RECOMMENDED)
fig,ax = plt.subplots()
ax.plot(x,[50,75,150,250]); # Add some Data
type(fig), type(ax)
```

### Matplotlib Example Workflow

```python
# 0) Import Matplotlib and get it ready to work in a jupyter notebook
%matplotlib inline
import matplotlib.pyplot as plt

# 1) Prepare Data
x = [1,2,3,4]
y = [11,20,37,41]

# 2) Setting up the Plot
fig,ax = plt.subplots(figsize = (5,5)) # ( Width, Height )

# 3) Plot axes
ax.plot(x,y)

# 4) Customising the plot
ax.set(title = "Simple Plot",
       xlabel = "X-Axis",
       ylabel = "Y-Axis" )

# 5) Saving the plot as an img
fig.savefig("C:/miniconda/MINICONDA_WORKSPACE/MatPlotLib_Section7/sample-plot.png")
```

### Using the normal distribution ( `np.random.randn(1000)` )

```python
x = np.random.randn(1000)
fig,ax = plt.subplots()
ax.hist(x);
```

### Methods to plot various plots side by side

```python
# Creating the various Subplots
fig , ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows = 2, ncols = 2)
ax1.plot(x,x/2);
ax2.hist(x)
ax3.scatter(np.random.random(10),np.random.random(10))
ax4.scatter(x,x**2);
```

```python
fig,ax = plt.subplots(nrows = 2, ncols = 2)
ax[0,0].plot(x,x/2);
ax[0,1].hist(x)
ax[1,0].scatter(np.random.random(10),np.random.random(10))
ax[1,1].scatter(x,x**2);
```
