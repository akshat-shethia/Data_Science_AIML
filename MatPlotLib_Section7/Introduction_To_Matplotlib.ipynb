### Which One Should You Use ( PyPlot vs MatPlotLib OO Method ? )
* When plotting something **quickly** ( just for **reference** ) Use the Pyplot method <br>
* When plotting something **advanced** Use the MatPlotLib OO method

```python
over_50 = heart_disease[heart_disease["age"] >50]
over_50.head()
```

```python
# PyPlot Method
over_50.plot(kind = "scatter", x = "age" , y = "chol", c = "target"); # c -> color
```

```python
## Object Oriented Method
fig,ax = plt.subplots(figsize = (10,6))
over_50.plot(kind = "scatter",x = "age", y = "chol", c = "target", ax = ax); #Only difference is that we have to add an ax parameter
# we can change the axis limits by using the following function
# ax.set_xlim([45,100]);
```

```python
## Object Oriented Method From Scratch
fig,ax = plt.subplots(figsize = (10,6))

# Creating a scatter plot
scatter = ax.scatter(x = over_50["age"], y = over_50["chol"], c = over_50["target"])

# Customizing the plot
ax.set(xlabel = "AGE", ylabel = "CHOLESTROL", title = "AGE VS CHOLESTROL SCATTER PLOT")

# Adding a legend
ax.legend(*scatter.legend_elements(),title = "Target");

# Adding a horizontal line
ax.axhline(over_50["chol"].mean(),linestyle = '--');
```

### Now combining age with max heart rate ( thalach )

```python
# Creating the plot using the OO method from scratch
fig,ax = plt.subplots(figsize = (10,6))

# Creating the actual plot
ax.scatter(x = over_50["age"], y = over_50["thalach"], c = over_50["thal"]);

# Customising the plot
ax.set(title = "AGE VS HEART RATE SCATTER PLOT", xlabel = "HEART RATE", ylabel = "AGE");

# Adding a horizontal line
ax.axhline(y = over_50["thalach"].mean(), linestyle = "--");
```

### Now combining both of these and creating both of them at the same time

```python
# Creating the basic subplots
fig,(ax1,ax2) = plt.subplots(nrows = 2, ncols = 1, figsize = (10,10),sharex = True) # Now both the graphs share the same x axis

# Creating a scatter plot for AGE VS CHOLESTROL
scatter = ax1.scatter(x = over_50["age"], y = over_50["chol"], c = over_50["target"])
ax1.set(xlabel = "AGE", ylabel = "CHOLESTROL", title = "AGE VS CHOLESTROL SCATTER PLOT")
ax1.legend(*scatter.legend_elements(),title = "Target");
ax1.axhline(over_50["chol"].mean(),linestyle = '--');

# Creating a scatter plot for AGE VS MAX HEART RATE
ax2.scatter(x = over_50["age"], y = over_50["thalach"], c = over_50["thal"]);
ax2.set(title = "AGE VS HEART RATE SCATTER PLOT", xlabel = "HEART RATE", ylabel = "AGE");
ax2.axhline(y = over_50["thalach"].mean(), linestyle = "--");

# Adding a title to the entire figure
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold");
```
